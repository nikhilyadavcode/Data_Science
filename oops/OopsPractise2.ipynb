```python
class Employee:
    def __init__(self,name,salary,department):
        self.name=name
        self.salary=salary
        self.department=department

    def display_info(self):
        print(f"Name:{self.name} Salary:{self.salary} department {self.department}")

    def annual_bonus(self):
        bonus=self.salary*0.10
        return bonus
    
    def promote(self,new_department,increment):
        self.department=new_department
        self.salary+=increment
        print(f"{self.salary} has been promoted to {self.department} with new salary {self.salary}")

    emp1=Employee("Mohit",70000,"CSE")
    emp1.display_info()
    print("Annual Bonus is:", emp1.annual_bonus())
    emp1.promote("Senior IT",10000)
    emp1.display_info()
```

```python
class Student:
    def __init__(self, name, salary, experience, skills):
        self.name = name
        self.base_salary = salary         # Salary attribute ka naam change kiya
        self.experience = experience
        self.skills = skills

    def calculate_salary(self):
        salary = self.base_salary

        # Experience ke basis par increment
        if self.experience >= 5:
            salary += 20000
        elif self.experience >= 2:
            salary += 10000

        # Skills ke basis par increment
        if "Python" in self.skills:
            salary += 5000
        if "Java" in self.skills:
            salary += 4000
        if "Data Science" in self.skills:
            salary += 15000

        return salary

    def display_info(self):
        total_salary = self.calculate_salary()
        print(f"Name: {self.name}")
        print(f"Base Salary: {self.base_salary}")
        print(f"Experience: {self.experience} years")
        print(f"Skills: {self.skills}")
        print(f"Final Salary: {total_salary}")
        

# Object
s1 = Student("Nikhil", 30000, 3, ["Python", "Java"])
s1.display_info()
s1 = Student("Mohit", 40000, 4, ["Python", "Java"])
s1.display_info()
s1 = Student("Aditya", 50000, 3, ["Data Science", "Java"])
s1.display_info()
```

```python
# Bank Account

class BankAccount:
    def __init__(self,initial_balance,pin):
        self.__balance=initial_balance
        self.__pin=pin


# Deposite Money
    def deposite(self,amount):
        if amount>0:
            self.__balance+=amount
            print("Deposite Successfully")
        else:
            print("Invalid Deposite Amount")

# Withdraw Money
    def withdraw(self,amount,pin):
        if pin!=self.__pin:
            print("Invalid pin")
            return
        
        if amount > self.__balance:
            print("Insufficient funds")
        else:
            self.__balance-=amount
            print("Withdrwal successfully!")

# check Balance
    def check_balance(self,pin):
        if pin==self.__pin:
            return self.__balance
        else:
            print("Invalid PIN")
            return None
    
    def add_interest(self,rate):
        self.__balance+=self.__balance*rate
        print("Interested added.")


account=BankAccount(10000,224176)
account.deposite(-500)
account.deposite(500)
print("Balance:",account.check_balance(224176))
account.withdraw(2000,224176)
print("Balance:",account.check_balance(224176))
account.withdraw(20000,201316)
# account.withdraw(17000,224176)
account.add_interest(0.06)
print("Balance:",account.check_balance(224176))
```

```python
# Mini Amazon Shopping system


class Product:
    # class variable: store all products in a shared catalog
    catalog = {}

    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock
        Product.catalog[name] = self  # add to global catalog

    def update_stock(self, quantity):
        """Update product stock"""
        self.stock += quantity

    @classmethod
    def show_catalog(cls):
        print("\n--- Product Catalog ---")
        for name, product in cls.catalog.items():
            print(f"{name} - ₹{product.price} ({product.stock} available)")


class ShoppingCart:
    def __init__(self, customer_name):
        self.customer_name = customer_name
        self.items = {}  # {product_name: quantity}

    def add_item(self, product_name, quantity=1):
        if product_name not in Product.catalog:
            print(f"Product '{product_name}' not found.")
            return
        product = Product.catalog[product_name]
        if quantity <= 0:
            print("Invalid quantity.")
            return
        if product.stock < quantity:
            print(f"Not enough stock for '{product_name}'. Only {product.stock} left.")
            return
        product.stock -= quantity
        self.items[product_name] = self.items.get(product_name, 0) + quantity
        print(f"Added {quantity} x '{product_name}' to cart.")

    def remove_item(self, product_name):
        if product_name in self.items:
            quantity = self.items.pop(product_name)
            Product.catalog[product_name].stock += quantity
            print(f"Removed '{product_name}' from cart.")
        else:
            print(f"'{product_name}' not in cart.")

    def view_cart(self):
        print(f"\n--- {self.customer_name}'s Cart ---")
        if not self.items:
            print("Cart is empty.")
            return
        for product_name, qty in self.items.items():
            price = Product.catalog[product_name].price
            print(f"{product_name}: {qty} x ₹{price} = ₹{qty * price}")

    def checkout(self, discount=0.0, tax_rate=0.18):
        """Calculate final amount with discount and tax"""
        if not self.items:
            print("Cart is empty. Cannot checkout.")
            return
        subtotal = sum(Product.catalog[name].price * qty for name, qty in self.items.items())
        discount_amount = subtotal * discount
        tax_amount = (subtotal - discount_amount) * tax_rate
        total = subtotal - discount_amount + tax_amount

        print(f"\nSubtotal: ₹{subtotal:.2f}")
        print(f"Discount: ₹{discount_amount:.2f}")
        print(f"Tax: ₹{tax_amount:.2f}")
        print(f"Total: ₹{total:.2f}")
        print(f"Thank you for shopping with us, {self.customer_name}!")
        self.items.clear()  # empty cart after checkout


# ------------------ USAGE ------------------

# Step 1: Create products (like Amazon adding inventory)
Product("iPhone 15", 79999, 10)
Product("MacBook Air", 119999, 5)
Product("AirPods", 15999, 20)
Product("Samsung S24", 84999, 8)

# Step 2: Show available products
Product.show_catalog()

# Step 3: Customer shopping flow
cart1 = ShoppingCart("Rahul")
cart1.add_item("iPhone 15", 1)
cart1.add_item("AirPods", 2)
cart1.view_cart()

# Step 4: Remove or adjust items
cart1.remove_item("AirPods")
cart1.view_cart()

# Step 5: Checkout with 10% discount
cart1.checkout(discount=0.10)

# Step 6: Display updated stock
Product.show_catalog()


```

```python
# Real World Example of Multi-level Inheritance
# Problem Statement:
# You want to create a Vehicle class hierarchy where you have a base Vehicle class, a Car class that inherits from Vehicle, and a SportsCar class that inherits from Car.

class Vehicle:
    def __init__(self, brand, model): # (brand="Toyota", model="Corolla")
        self.brand = brand # self.brand = Toyota
        self.model = model # self.model = Corolla

    def show_details(self):
        print(f"Vehicle Brand: {self.brand}, Model: {self.model}")

class Car(Vehicle):
    def __init__(self, brand, model, doors): # (brand="Honda", model="Civic", doors=4)
        super().__init__(brand, model)  # Call parent constructor
        self.doors = doors

    def show_details(self):
        super().show_details()  # Call parent method
        print(f"Number of Doors: {self.doors}")

class SportsCar(Car):
    def __init__(self, brand, model, doors, top_speed): # (brand="Ferrari", model="488", doors=2, top_speed=211)
        super().__init__(brand, model, doors)  # Call parent constructor
        self.top_speed = top_speed

    def show_details(self):
        super().show_details()  # Call parent method
        print(f"Top Speed: {self.top_speed} mph")

# Using the classes
vehicle = Vehicle("Toyota", "Corolla")
vehicle.show_details()

car = Car("Honda", "Civic", 4)
car.show_details()

sports_car = SportsCar("Ferrari", "488", 2, 211)
sports_car.show_details()
```

```python
# 3. Hierarchical Inheritance
# https://media.geeksforgeeks.org/wp-content/uploads/20200108144949/Hierarchical-inheritance1.png
# Concept:
# Multiple child classes inherit from a single parent class.

# Simple Example of Hierarchical Inheritance
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

class Cat(Animal):
    def meow(self):
        print("Cat meows")

class Cow(Animal):
    def moo(self):
        print("Cow moos")

dog = Dog()
dog.bark()   # Dog barks
dog.speak()  # Animal speaks

cat = Cat()
cat.meow()   # Cat meows
cat.speak()  # Animal speaks

cow = Cow()
cow.moo()    # Cow moos
cow.speak()  # Animal speaks
# cow.bark()   # Error: 'Cow' object has no attribute 'bark'
```

```python
# Real World Example of Hierarchical Inheritance
# Problem Statement:
# You want to create a simple content management system where you have a base Content class and multiple child classes like Article, Video, and Podcast that inherit from Content.

class Content:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def display_info(self):
        print(f"Title: {self.title}, Author: {self.author}")

class Article(Content):
    def __init__(self, title, author, word_count):
        super().__init__(title, author)
        self.word_count = word_count

    def display_info(self):
        super().display_info()
        print(f"Word Count: {self.word_count}")

class Video(Content):
    def __init__(self, title, author, duration):
        super().__init__(title, author)
        self.duration = duration

    def display_info(self):
        super().display_info()
        print(f"Duration: {self.duration}")

class Podcast(Content):
    def __init__(self, title, author, length):
        super().__init__(title, author)
        self.length = length

    def display_info(self):
        super().display_info()
        print(f"Length: {self.length}")


# Using the classes
article = Article("OOP in Python", "Alice", 1500)
article.display_info()

video = Video("Python Tutorial", "Bob", 10)
video.display_info()

podcast = Podcast("Tech Talks", "Charlie", 30)
podcast.display_info()
```
