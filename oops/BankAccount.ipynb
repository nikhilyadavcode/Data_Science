```python
# Bank Account

class BankAccount:
    def __init__(self,initial_balance,pin):
        self.__balance=initial_balance
        self.__pin=pin


# Deposite Money
    def deposite(self,amount):
        if amount>0:
            self.__balance+=amount
            print("Deposite Successfully")
        else:
            print("Invalid Deposite Amount")

# Withdraw Money
    def withdraw(self,amount,pin):
        if pin!=self.__pin:
            print("Invalid pin")
            return
        
        if amount > self.__balance:
            print("Insufficient funds")
        else:
            self.__balance-=amount
            print("Withdrwal successfully!")

# check Balance
    def check_balance(self,pin):
        if pin==self.__pin:
            return self.__balance
        else:
            print("Invalid PIN")
            return None
    
    def add_interest(self,rate):
        self.__balance+=self.__balance*rate
        print("Interested added.")


account=BankAccount(10000,224176)
account.deposite(-500)
account.deposite(500)
print("Balance:",account.check_balance(224176))
account.withdraw(2000,224176)
print("Balance:",account.check_balance(224176))
account.withdraw(20000,201316)
# account.withdraw(17000,224176)
account.add_interest(0.06)
print("Balance:",account.check_balance(224176))
```
