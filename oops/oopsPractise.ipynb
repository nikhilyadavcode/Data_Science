```python
# class method:-
# A class method is bound to the class not the object(instance).
# it can access and modify class-level variables.
# it is defined using the @classmethod decorator.
# The First paramater is always cls,which represent the class.

# Syntax
# @classmethod
# def method_name(cls,args):
# code
```

```python
class Student:
    school_name="ABC school"

    def __init__(self,name):
        self.name=name

    @classmethod
    def change_school(cls,new_name):
        cls.student_name=new_name

Student.change_school("XYZ school")
print(Student.school_name)
```

```python
# why we need a class method?
# Normally jab hm class ke ander method banate hai (without any decorator) toh wo object 
# ke liye kaam karta hai matlab usake paas self hota hai jo instance variables ko acces karta hai

# Example
class student:
    def __init__(self,name):
        self.name=name

    def show(self):
        print("Student name:",self.name)

s1=student("Nikhil")
s1.show()
```

```python
# Need:-
# jab hum class-level data (not object data) change karna chahein
class student:
    school_name="Anand Memorial Academy"

    @classmethod
    def change_school(cls,new_name):
        cls.school_name=new_name

student.change_school("Gautam Buddha University")
print(student.school_name)
```

```python
# Need:-
# Factory methods banane ke liye

class Date:
    def __init__(self,day,month,year):
        self.day=day
        self.month=month
        self.year=year

    @classmethod
    def from_string(cls,date_str):
        d,m,y=map(int,date_str.split('-'))
        return cls(d,m,y)
    
date1=Date.from_string("02-11-2025")
print(date1.day,date1.month,date1.year)
```

```python
# Need:-
# Jab class aur subclass dono me common setup karna
# Agar hum class inheritance use kar rahe hain (parent-child relationship),
# toh class method subclass ke liye bhi automatically adjust ho jate hain

class Animal:
    species="unknown"

    @classmethod
    def set_species(cls,name):
        cls.species=name

class Dog(Animal):
    pass

Dog.set_species("Canine")
print(Dog.species)
```
