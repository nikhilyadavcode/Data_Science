```python
class Car:
    def __init__(self,color,model):
        self.model=model
        self.color=color
        self.speed=0

    def start(self):
        print(f"{self.model} started.")

    def accelerate(self):
        self.speed+=10
        print(f"{self.speed} is now at {self.speed} km/h")


my_car=Car("Red","Tasala Model 3")

print(my_car.color)
my_car.start()
my_car.accelerate()
```

```python
# using constructor method
class Student:
    def __init__(self,n,a,p,c,m):
        self.name=n
        self.age=a
        self.physics=p
        self.chemistry=c
        self.math=m

    def total_marks(self):
        total=self.physics+self.chemistry+self.math
        return total
    
    def calculate_percentage(self):
        total=self.total_marks()
        percentage=(total/300)*100
        return percentage
    
s1=Student("Nikhil",23,85,87,92)
print("total marks scored by",s1.name,"is",s1.total_marks())
print("Percentage scored by",s1.name,"is",s1.calculate_percentage(),"%")
s2 = Student("Rahul", 22, 75, 80, 70)
print("Total marks scored by", s2.name, "is", s2.total_marks())
print("Percentage scored by", s2.name, "is", s2.calculate_percentage(), "%")
```
