```python
# Find the quartile
import numpy as np
data=[2,4,4,5,6,7,8]
Q1=np.percentile(data,25)
Q2=np.percentile(data,50)
Q3=np.percentile(data,75)

print("Q1=",Q1)
print("Q2=",Q2)
print("Q3=",Q3)
```

```python
# standard Deviation
import math
data=[600,470,170,430,300]
n=len(data)
mean=sum(data)/n
variance=sum((x-mean)**2 for x in data)/n
std_dev=math.sqrt(variance)
print("Variance is :",variance)
print("Standard Deviation:",std_dev)
```

```python
import numpy as np
data=[600,470,170,430,300]
print("Population standard deviation",np.std(data))
print("Sample Standard deviation",np.std(data,ddof=1))
```

```python
# mean deviation
import math
data=[600,470,170,430,300]
n=len(data)
mean=sum(data)/n
mean_dev=sum(abs(x-mean) for x in data)/n
print("Mean Deviation are:",mean_dev)
```

```python
import numpy as np
data=np.array([600,470,170,430,300])
mean=np.mean(data)
mean_deviation=np.mean(np.abs(data-mean))
print("Mean Deviation=",mean_deviation)
```

```python
# z score
import math

data = [600, 470, 170, 430, 300]
n = len(data)

mean = sum(data) / n

variance = sum((x - mean) ** 2 for x in data) / n
std_dev = math.sqrt(variance)

z_scores = [(x - mean) / std_dev for x in data]

print("Mean:", mean)
print("Standard Deviation:", std_dev)
print("Z-scores:", z_scores)
```
