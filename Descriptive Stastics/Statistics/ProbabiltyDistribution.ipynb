```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
# Normal Distribution>>gaussin distribution>>majority of data points are clustered around mean

mu=0
sigma=0.1

s=np.random.normal(mu,sigma,size=1000)
len(s)
```

```python
plt.hist(s,30)
plt.show()
```

```python
sns.distplot(s)
```

```python
# uniform distribution>>all data outcomes are equally probability to occur in a range

u=np.random.uniform(0,1,10000)
plt.hist(u)
plt.show()
```

```python
sns.displot(u)
```

```python
sns.distplot(u)
```

```python
# log normal distribution 
l=np.random.lognormal(mean=0, sigma=0.5, size=1000)
l=l[np.where(abs(1)<5)]
plt.hist(1,bins=100)
plt.show()
```

```python
sns.boxplot(l) # since lognormal is right skewed outliers will be there
```

```python
sns.distplot(u)
```

```python
# log normal>>normal distribution
sns.displot(np.log(l))
```

```python
# square root transformation
sns.distplot(np.sqrt(l))
```

```python
# chi square distribution
c=np.random.chisquare(df=30, size=10000)
c=c[np.where(abs(c)<100)]
plt.hist(c,bins=1000)
plt.show()
```

```python
c=np.random.chisquare(df=30, size=10000)
d=np.random.chisquare(df=10, size=10000)
c=c[np.where(abs(c)<100)]
d=d[np.where(abs(d)<100)]
plt.hist(c,bins=1000)
plt.hist(d,bins=1000)
plt.show()
```

```python
# t-distribution
t=np.random.standard_t(df=3, size=10000)
t=t[np.where(abs(t)<5)]
plt.hist(t,bins=1000)
plt.show()
```

```python
# f distribution
f=np.random.f(dfnum=5, dfden=2, size=10000)
f=f[np.where(abs(f)<5)]
plt.hist(f,bins=1000)
plt.show()
```

```python
# discrete distribution
# binomial distribution>>bionomial distribution one trial is bionomial dist
b=np.random.binomial(n=10, p=0.5, size=1000)
plt.hist(b,bins=30)
plt.show()
```

```python
# binomial distribution with n=100
b=np.random.binomial(n=100, p=0.5, size=1000)
plt.hist(b)
plt.show()
```

```python
# poisson distribution
p=np.random.poisson(lam=5, size=1000)
plt.hist(p,bins=30)
plt.show()
```
