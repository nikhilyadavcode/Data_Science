```python
# Data interpolation>> is a process of estimating unknown values tha known values

# linear interpolation
# cubic
# polynomial interpolation
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5])
y=np.array([10,15,14,10,18])
plt.scatter(x,y)
```

```python
# linear interpolation >> linear in nature
x_new=np.linspace(1,5,50) # create new x value
y_new=np.interp(x_new,x,y) # linear interpolation
plt.plot(x_new,y_new)
```

```python
plt.scatter(x,y)
```

```python
# cubic interpolation >> 
x=np.array([1,2,3,4,5])
y=np.array([10,15,14,10,18])
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d
f=interp1d(x,y,kind='cubic') 
```

```python
f
```

```python
x_new=np.linspace(1,5,50)
y_interpolated=f(x_new)
```

```python
y_interpolated
```

```python
plt.scatter(x_new,y_interpolated)
```

```python
# polynomial interpolation
x=np.array([1,2,3,4,5])
y=np.array([10,15,14,10,18])
p=np.polyfit(x,y,deg=3)  # degree 3 polynomial fit
```

```python
p=np.polyfit(x,y,deg=3) 
x_new=np.linspace(x,y,3)
y_poly=np.polyval(p,x_new)
plt.scatter(x_new,y_poly)
```

```python
plt.scatter(x,y)
```
