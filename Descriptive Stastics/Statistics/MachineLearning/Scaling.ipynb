```python
import pandas as pd
import numpy as np
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# features scaling
# standardization >> mu=0,sigma=1
# normalization(minmax)>> data comes between 0 ans 1
# unit vector >> 1
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
sns.displot(df['total_bill'])
```

```python
mean=np.mean(df['total_bill'])
std=np.std(df['total_bill'])
```

```python
mean
```

```python
std
```

```python
std_data=[]
for i in list(df['total_bill']):
    z=(i-mean)/std
    std_data.append(z)
std_data
```

```python
# feature scaling is option it doesn't change the distribution of data
```

```python
sns.displot(df['total_bill'],kde=True)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler
```

```python
scaler.fit(df[['total_bill']]) # it will calculate mean and sigma
```

```python
scaler
```

```python
scaler.transform(df[['total_bill']])
```

```python
df
```

```python
scaler.fit_transform(df[['total_bill',"tip"]])
```

```python
pd.DataFrame(scaler.fit_transform(df[['total_bill',"tip"]]), columns=['total_bill','tip'])
```

```python
scaler.transform([[13,12]])
```

```python
# normalization/minmax scaling
from sklearn.preprocessing import MinMaxScaler
min_max=MinMaxScaler()
```

```python
min_max.fit_transform(df[['total_bill',"tip"]])
```

```python
min_max.transform([[13,12]])
```
