```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["Features"] = df.columns
vif["VIF"] = [variance_inflation_factor(df.values, i) 
              for i in range(len(df.columns))]

print(vif)
```

```python

# id vif>10 >> drop the feature one by one>> here highest is longitude>> we will drop this and recalculate VIF again
# calculate VIF again after dropping the highest one
df1=df.copy()
```

```python
df1
```

```python
df1.drop('Longitude',axis=1,inplace=True)
```

```python
vif = pd.DataFrame()
vif["Features"] = df1.columns
vif["VIF"] = [variance_inflation_factor(df1.values, i) 
              for i in range(len(df1.columns))]
```

```python
vif
```

```python
df1.drop('AveRooms',axis=1,inplace=True)
vif = pd.DataFrame()
vif["Features"] = df1.columns
vif["VIF"] = [variance_inflation_factor(df1.values, i) 
              for i in range(len(df1.columns))]
```

```python
vif
```

```python
df1.drop('Latitude',axis=1,inplace=True)
vif = pd.DataFrame()
vif["Features"] = df1.columns
vif["VIF"] = [variance_inflation_factor(df1.values, i) 
              for i in range(len(df1.columns))]
```

```python
X=df1.iloc[:,:-1] # independent variables
y=df1.iloc[:,-1]  # dependent variable
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test =train_test_split(X,y,test_size=0.30,random_state=1)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
# scaling (optional) can be done after train test split
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler
```

```python
# fit_transform is used for train data
# transform is used for test data
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
X_train
X_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model
```

```python
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
# performance metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
mae=mean_absolute_error(y_test,y_pred)
mae
```

```python
print("Mean Absolute Error:",mae)
mse=mean_squared_error(y_test,y_pred)
print("Mean Squared Error:",mse)
r2=r2_score(y_test,y_pred)
print("R2 Score:",r2)

```
