```python
# Generator Function:-A generator function is a special type of function in python that
# returns a sequence of values,one at a time using the yield keyword instead of return

# syntax:-
def generator():
    yield 1
    yield 2
    yield 3
    
```

```python
def number():
    yield 1
    yield 2
    yield 3

gen=number()
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
def count_up_to(n):
    for i in range(1,n+1):
        yield i

for num in count_up_to(5):
    print(num)
```

```python
def even_number(n):
    for i in range(1,n+1):
        if i%2==0:
         yield i

for num in even_number(20):
    print(num)
```

```python
def even_number(limit):
    for i in range(2,limit+1,2):
        yield i

evens=even_number(10)

for e in evens:
    print(e)
```

```python
#ðŸ”¹ Advantages of Generator Functions

#Memory Efficient: Doesnâ€™t store all values in memory.

#Lazy Evaluation: Values are generated only when needed.

#Infinite Sequences Possible: Like infinite Fibonacci or natural numbers.

#Readable and Easy to Write: Simple code for iterators.
```

```python
def fibonacci():
    a,b=0,1
    while True:
        yield a
        a,b=b,a+b

gen=fibonacci()

for i in range(10):
    print(next(gen))
```

```python
def generator_number():
    i=1
    while True:
        yield i
        i+=1

gen=generator_number()
for i in range(10):
    print(next(gen))
```

```python
def slogans():
    yield "vote for development (BJP)"
    yield "Har Ghar Congress(INC)"
    yield "Sabka Hoga Vikas(AAP)"
for slow in slogans():
        print(slow)
```

```python
def candidates_names(names):
    for n in names:
        yield n

candidates_names=["Rahul Gandhi","Narendra Modi","Arvind kejriwal"]

for name in candidates_names:
    print(name)
    
```

```python
# map function:-map() function is used to apply a function to every element of a list (or any iterable)
# syntax map(function,iterable)

def square(x):
    return x*x

numbers=[1,2,3,4,5,6]
result=map(square,numbers)
print(list(result))
```

```python
# reduce():-reduce() function is used to apply a function repeatedlly to all element of a list
# (or iterable).
# it reduce the list to a single value (like sum,product,max etc)

# syntax:
# from functools import reduce
# reduce(function,iterable)

from functools import reduce
numbers=[1,2,3,4,5,6]
result=reduce(lambda a,b:a+b,numbers)
print(result)
```

```python
### Task 1 â€“ Basic List Comprehension

#You are analyzing â€œPoha Chivdaâ€ packets sold across 5 cities.
#Each packet contains weight in grams: `[95, 105, 110, 85, 120]`
#Use a list comprehension to convert these to kilograms.

weights_grams=[95,105,110,85,120]
weight_Kg=[w/1000 for w in weights_grams]
print(weight_Kg)
```

```python
### Task 2 â€“ Conditional List Comprehension

#Given a list of prices for â€œCornflakes Chivdaâ€ packets: `[45, 60, 75, 30, 90]`,
#use a list comprehension to select only those packets priced above â‚¹50.

prices=[45,60,75,30,90]
above_50=[p for p in prices if p>50]
print(above_50)
```

```python
### Task 3 â€“ Nested List Comprehension

#A supplier sends the daily sales data for 3 Chivda types:
#[[10, 15, 20], [5, 10, 5], [8, 12, 16]]`
#Use a nested list comprehension to create a flat list of all quantities.


sales_data=[[10,15,20],[5,10,5],[8,12,16]]
flat_list=[qty for sublist in sales_data for qty in sublist ]
print(flat_list)
```

```python
### Task 5 â€“ Filter Function

#For â€œDiet Chivdaâ€, you want to filter out packets with more than 150 calories.
#Use `filter()` with a lambda to select only those â‰¤150 from `[100, 180, 120, 90, 160]`.

items=filter(lambda a:a>=150,[100,180,120,90,160])
print(list(items))
```

```python

#Combine `map()` and `filter()` to first increase all prices of â€œNamakpara Chivdaâ€ by â‚¹10,
#and then filter only those that cost more than â‚¹50.
#Original list: `[30, 45, 60, 25, 70]`

prices=[30,45,60,25,70]
updated_prices=filter(lambda a:a>50,map(lambda a:a+10,prices))
print(list(updated_prices))
```

```python
### Task 8 â€“ Conditional Comprehension with Text

#Given Chivda names: `["poha", "cornflakes", "masala", "diet", "navratna"]`,
#use a list comprehension to create a new list of only those names containing the letter `"a"`.

names=["poha","cornflask","masala","diet","navratna"]
names_with_a=[name for name in names if 'a' in name]
print(names_with_a)
```

```python
### Task 9 â€“ Function + Lambda Combo

#Define a function `apply_offer(prices, discount)` that takes a list of prices and uses a lambda inside a `map()` to apply the given discount percentage.
#Example: `apply_offer([100, 200, 300], 20)` â†’ `[80.0, 160.0, 240.0]`

def apply_offer(prices,discount):
    return list(map(lambda a:a-(a*discount/100),prices))

print(apply_offer([100, 200, 300], 20))
```
