```python
# adaboost classifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

import warnings
warnings.filterwarnings('ignore')
```

```python
X,y=make_classification(n_samples=1000,n_features=20,n_classes=2,random_state=1)
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
classifier=AdaBoostClassifier()
classifier
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
y_pred
```

```python
# Evaluate the model
print("Current model performance:")
print("Accuracy:{accuracy_score(y_test,y_pred)}")
print("Classification Report:")
print(classification_report(y_test,y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={
    'n_estimators':[50,100,200],
    'learning_rate':[0.01,0.1,1.0,1.5,2.0],
    'algorithm':['SAMME']
}
```

```python
ada=AdaBoostClassifier()
grid_search=GridSearchCV(estimator=ada,param_grid=param_grid,cv=5,verbose=3)
grid_search
```

```python
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
best_model=grid_search.best_estimator_
```

```python
y_pred=best_model.predict(X_test)

# Evaluate the best model
print("Current model performance:")
print("Accuracy:{accuracy_score(y_test,y_pred)}")
print("Classification Report:")
print(classification_report(y_test,y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test,y_pred))
```

```python
# Adaboost Regression
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
X,y=make_regression(n_samples=1000,n_features=10,random_state=1)
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
regressor=AdaBoostRegressor()
regressor
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)

# Evaluate the model
print("Current model performance:")
print(f"R2 Score:{r2_score(y_test,y_pred)}")
print(f"mean_absolute_error:{mean_absolute_error(y_test,y_pred)}")
print(f"mean_squared_error:{mean_squared_error(y_test,y_pred)}")

```

```python
# Hyperparameter tuning
from sklearn.model_selection import GridSearchCV
param_grid={
    'n_estimators':[50,100,200],
    'learning_rate':[0.01,0.1,0.5,1.0,1.5],
    'loss':['linear','square','exponential']
}
grid_search=GridSearchCV(estimator=regressor,param_grid=param_grid,cv=5,verbose=3)
grid_search
```

```python
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
best_model=grid_search.best_estimator_
y_pred_tuned=best_model.predict(X_test)

# Evaluate the tuned model
print("Tuned Model performance:")
print(f"R2 score:{r2_score(y_test,y_pred_tuned)}")
print(f"Mean Absolute Error:{mean_absolute_error(y_test,y_pred_tuned)}")
print(f"Mean Squared Error:{mean_squared_error(y_test,y_pred_tuned)}")
```
