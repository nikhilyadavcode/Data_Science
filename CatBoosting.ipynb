```python
# catboost classifier

from catboost import CatBoostClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
# Generate datasets

X,y=make_classification(n_samples=1000,n_features=20,n_classes=2,random_state=1)
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=1)
```

```python
X_train.shape,X_test.shape
```

```python
classifier=CatBoostClassifier(verbose=0) # if you don't want details in training put verbose=0
classifier
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)

# Evaluate the model
print("Current model performance:")
print(f"Accuracy:{accuracy_score(y_test,y_pred)}")
print("Classification Report:")
print(classification_report(y_test,y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test,y_pred))
```

```python
# Hyperparameter tuning using GridSearchCV

param_grid={
    'iterations':[100,200,300],
    'learning_rate':[0.01,0.05,0.1,0.2],
    'depth':[3,4,5,6]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid_search=GridSearchCV(estimator=CatBoostClassifier(verbose=0),param_grid=param_grid,cv=5,n_jobs=-1,verbose=3)
grid_search.fit(X_train,y_train)
```

```python
# Get best parameters

print("Best Parameters:",grid_search.best_params_)

# Train best model

best_model=grid_search.best_estimator_
y_pred_tuned=best_model.predict(X_test)

print("Tuned model Performance:")
print(f"Accuracy:{accuracy_score(y_test,y_pred_tuned)}")
print("Classification Report:")
print(classification_report(y_test,y_pred_tuned))
print("Confusion Matrix:")
print(confusion_matrix(y_test,y_pred_tuned))
```

```python
# Catboost regressor
from catboost import CatBoostRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
X,y=make_regression(n_samples=1000,n_features=2,noise=10,random_state=1)
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=1)
```

```python
X_train.shape,X_test.shape
```

```python
regressor=CatBoostRegressor(verbose=0)
regressor
```

```python
regressor.fit(X_train,y_train)
```

```python
# Predict
y_pred=regressor.predict(X_test)

# Evaluate the model
print('Current model performance:')
print(f"R2 score{r2_score(y_test,y_pred)}")
print(f"Mean absolute Error:{mean_absolute_error(y_test,y_pred)}")
print(f"Mean Squared Error:{mean_squared_error(y_test,y_pred)}")
```

```python
# Hyperparameter tuning using GridSearchCV
param_grid={
    'iterations':[100,200,300],
    'learning_rate':[0.01,0.05,0.1,0.2],
    'death':[3,4,5,6]
}
```

```python
grid_search = GridSearchCV(
    estimator=CatBoostClassifier(verbose=0),
    param_grid=param_grid,
    cv=5,
    n_jobs=-1,
    verbose=2
)

grid_search.fit(X_train, y_train)
```
