```python
# Using print statement to debug
def divide(a,b):
    return a/b

for i in range(-10,10):
    print(i,i+1)
    divide(i,i+1)
```

```python
# Using another print statement inside function
def divide(a,b):
    print(a,b)
    return a/b

for i in range(-10,10):
    divide(i,i+1)
```

```python
# from print statement you understand that when b is 0,it throws the error

def divide(a,b):
    print(a,b)
# Breaking the code in terms of execution
    if b==0:
        print("Zero division error")
        return
    return a/b

for i in range(-10,10):
    # print(i,i+1)
    divide(i,i+1)
```

```python
# NOSET level is not used
# 1.DEBUG>>lowest level of logging.To give detailed information of any variable,msg,while doing debugging.
# 2.info>> This is used to convey that the code is working as excepted.
# example "Data analysis is completed"
# 3.WARNING >> used to indicate that something unexpected happened or potential issue in the code.
# 4.ERROR>>Serious problem with some function.
# 5 CRITICAL>>Termination of program/code/software.

# the above level of logging is executed in the same order>> debug being lowest priority and critical being highest priority.
# while defining level in logging.basicConfig,from that level the logs will be logged or in the others words the log level priority will start from that level which is used to initialise the logging 

# example>> if you initialise with WARNING then only,error,critical msg will be logged,debug and info will not be logged.
```

```python
import logging
logging.basicConfig(filename="test_new.log",level=logging.DEBUG,format='%(asctime)s %(message)s')
```

```python
logging.debug("This msg is for debugging")
logging.info("This is my info msg")
logging.warning("This is my warning msg")
logging.shutdown()
```

```python
# write a program to seprate the integer and string in two lists seprately.
l=[1,"hello",[2,"world"],4,[3,"python"]]
print(l)
```

```python
l1_int=[]
l2_str=[]

for i in l:
    logging.info(f"processing each element {i}")

    if type(i)==list:
        for j in i:
            logging.info(f"processing sublist element:{j}")
            if type(j)==int:
                l1_int.append(j)
    elif type(i)==int:
        l1_int.append(i)
    else:
        l2_str.append(i)
logging.info(f"The result is:{l1_int},{l2_str}")
logging.shutdown()
```
