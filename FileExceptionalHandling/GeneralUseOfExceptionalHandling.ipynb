```python
# Zero division error doesnt make any sense here
def test(file):
    try:
        f=open(file,"r")
        f.read()
    except FileNotFoundError as e:
        print(f"My file was not found {e}")
    except ZeroDivisionError as e:
        print("The division is not possible due the error>>",e)

test("Nikhil.txt")
```

```python
# best practise::give proper documentation.proper comments in your code.

def test(file):
    """
  This function is about is reading a file.

  Args:
  input:file(str)
  output:text(str)
    """
    try:
       f=open(file,"r")
       f.read()
    except FileNotFoundError as e:
        print(f"My file was not found {e}")
    # unnecesary exceptation
    except ZeroDivisionError as e:
        print("The division is not possible due to error>>",e)

test("Nikhil.txt")
```

```python
# best practise >> always clean up the resources using finally

def test(file):

    try:
        f=open(file,"rw")
        f.read()
    except FileNotFoundError as e:
        print(f"My file was not found {e}")
    finally:
        f.close()

test("Nikhil.txt")
```

```python
def test(file):

    try:
        f=open(file,"w")
        f.write("Anything being written")
    except FileNotFoundError as e:
        print(f"My file was not found {e}")
    finally:
        f.close()

test("Nikhil.txt")
```
